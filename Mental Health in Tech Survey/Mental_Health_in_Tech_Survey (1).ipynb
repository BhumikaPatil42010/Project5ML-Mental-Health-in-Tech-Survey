# **Mental Health in Tech Survey**

The “Mental Health in Tech Survey” project focuses on analyzing the prevalence of mental health issues and workplace attitudes toward mental health within the technology industry. The dataset, collected in 2014 by Open Sourcing Mental Illness, contains responses from employees across different countries, primarily working in tech environments. The main objective of this project is to explore how factors such as work environment, company policies, and personal background influence mental health awareness and treatment among employees.

The analysis begins with data preprocessing and cleaning, which is a crucial step in any data analysis task. The dataset contains missing values, inconsistent entries (especially in columns like gender), and irrelevant features such as comments and state (for non-US respondents). These issues are handled by removing unnecessary columns, filling missing values with appropriate substitutes, and standardizing categorical variables. For example, gender entries are normalized into consistent categories such as male, female, and others, while unrealistic age values are filtered out to maintain data quality.

Following data cleaning, exploratory data analysis (EDA) is performed to uncover patterns and trends in the dataset. Univariate analysis helps in understanding the distribution of individual variables such as age, gender, and treatment status. It reveals that a significant portion of respondents have sought treatment for mental health conditions, indicating the growing awareness of mental health issues in the tech industry.

Bivariate and multivariate analyses are then conducted to examine relationships between variables. One of the key findings is that individuals with a family history of mental illness are more likely to seek treatment. Additionally, employees who report that mental health interferes with their work are also more inclined to pursue professional help. The study also highlights that workplace factors play a significant role in shaping mental health outcomes. For instance, employees working in companies that provide mental health benefits and wellness programs are more likely to seek support.

Geographical analysis shows variations in responses across countries, suggesting that cultural and regional differences influence attitudes toward mental health. Another important observation is that many employees are hesitant to discuss mental health issues openly with coworkers, supervisors, or during job interviews due to fear of negative consequences. This indicates the presence of stigma surrounding mental health in professional settings.

The project also compares perceptions of mental health versus physical health in the workplace. The results suggest that mental health is often not treated with the same level of seriousness as physical health, which further emphasizes the need for improved awareness and supportive workplace policies.

In conclusion, this project provides valuable insights into the state of mental health in the tech industry. While awareness and treatment-seeking behavior have improved over time, there is still a significant gap in workplace support and openness. Organizations need to implement better mental health policies, promote awareness programs, and create a safe environment where employees feel comfortable discussing their mental health concerns. This analysis highlights the importance of data-driven decision-making in addressing critical issues like mental health in the workplace.

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# ================================
# DATA WRANGLING / DATA CLEANING
# ================================

# Create a copy to preserve original data
df_clean = df.copy()

# -------------------------------
# 1. Drop unnecessary columns
# -------------------------------
# 'comments' has too many missing values and is not useful for analysis
# 'state' is mostly null for non-US respondents

df_clean.drop(['comments', 'state'], axis=1, inplace=True, errors='ignore')


# -------------------------------
# 2. Handle Missing Values
# -------------------------------

# Fill categorical missing values with appropriate labels
df_clean['self_employed'].fillna('No', inplace=True)
df_clean['work_interfere'].fillna("Don't know", inplace=True)

# For other columns, fill with mode (most frequent value)
for col in df_clean.columns:
    if df_clean[col].dtype == 'object':
        df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)


# -------------------------------
# 3. Clean Gender Column
# -------------------------------

# Convert to lowercase
df_clean['Gender'] = df_clean['Gender'].str.lower().str.strip()

# Standardize gender values
male_terms = ['male', 'm', 'man', 'male-ish', 'maile', 'mal', 'msle']
female_terms = ['female', 'f', 'woman', 'femake', 'cis-female']

df_clean['Gender'] = df_clean['Gender'].replace(male_terms, 'male')
df_clean['Gender'] = df_clean['Gender'].replace(female_terms, 'female')

# Remaining values → 'others'
df_clean['Gender'] = df_clean['Gender'].apply(
    lambda x: x if x in ['male', 'female'] else 'others'
)


# -------------------------------
# 4. Clean Age Column
# -------------------------------

# Remove unrealistic age values
df_clean = df_clean[(df_clean['Age'] >= 15) & (df_clean['Age'] <= 60)]


# -------------------------------
# 5. Standardize Country Names
# -------------------------------

# Convert to lowercase
df_clean['Country'] = df_clean['Country'].str.lower().str.strip()

# Replace common variations
df_clean['Country'] = df_clean['Country'].replace({
    'united states': 'USA',
    'united states of america': 'USA',
    'us': 'USA',
    'uk': 'UK'
})


# -------------------------------
# 6. Remove Duplicates
# -------------------------------

df_clean.drop_duplicates(inplace=True)


# -------------------------------
# 7. Final Check
# -------------------------------

print("Final Shape:", df_clean.shape)
print("\nMissing Values After Cleaning:\n", df_clean.isnull().sum())
```

### What all manipulations have you done and insights you found?

# Data Manipulations Performed
1.Removed Irrelevant Columns - Dropped comments and state

Reason: High missing values and low analytical importance

'No', "Don't know", or mode values

3.Standardized Gender Column - Converted multiple inconsistent entries into:
